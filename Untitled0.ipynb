```python
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.colors as mcolors
import random

# Constants
EMPTY = 0
RED = 1
BLUE = 2

# Parameters
GRID_SIZE = 30
RATIO_EMPTY = 0.1
SIMILARITY_THRESHOLD = 0.3
MAX_STEPS = 100

# Create the grid
def create_initial_grid(size, empty_ratio):
    total_cells = size * size
    num_empty = int(total_cells * empty_ratio)
    num_agents = total_cells - num_empty
    num_each = num_agents // 2

    cells = [EMPTY] * num_empty + [RED] * num_each + [BLUE] * (num_agents - num_each)
    random.shuffle(cells)
    return np.array(cells).reshape((size, size))

# Plotting
def plot_grid(grid):
    cmap = mcolors.ListedColormap(['white', 'red', 'blue'])
    plt.figure(figsize=(6, 6))
    plt.imshow(grid, cmap=cmap, interpolation='nearest')
    plt.axis('off')
    plt.tight_layout()
    plt.show()

# Check satisfaction of agent
def is_satisfied(grid, x, y, threshold):
    agent = grid[x, y]
    if agent == EMPTY:
        return True

    neighbors = []
    for dx in [-1, 0, 1]:
        for dy in [-1, 0, 1]:
            if dx == 0 and dy == 0:
                continue
            nx, ny = x + dx, y + dy
            if 0 <= nx < grid.shape[0] and 0 <= ny < grid.shape[1]:
                neighbor = grid[nx, ny]
                if neighbor != EMPTY:
                    neighbors.append(neighbor)

    if not neighbors:
        return True

    similar = sum(1 for n in neighbors if n == agent)
    return (similar / len(neighbors)) >= threshold

# Perform one step
def step(grid, threshold):
    unsatisfied = []
    empties = list(zip(*np.where(grid == EMPTY)))

    for x in range(grid.shape[0]):
        for y in range(grid.shape[1]):
            if grid[x, y] != EMPTY and not is_satisfied(grid, x, y, threshold):
                unsatisfied.append((x, y))

    random.shuffle(unsatisfied)
    for x, y in unsatisfied:
        if empties:
            new_pos = random.choice(empties)
            empties.remove(new_pos)
            grid[new_pos], grid[x, y] = grid[x, y], EMPTY

    return grid

# Full simulation
def simulate(threshold, max_steps=100):
    grid = create_initial_grid(GRID_SIZE, RATIO_EMPTY)
    for _ in range(max_steps):
        grid = step(grid, threshold)
    return grid

# Run and plot
final_grid = simulate(SIMILARITY_THRESHOLD, MAX_STEPS)
plot_grid(final_grid)
```
